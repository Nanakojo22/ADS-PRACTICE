# IBM EMPLOYEE ATTRITION ANALYSIS

```python
#importing modules
import pandas as pd
```

```python
ibm_attrition=pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
ibm_attrition.head()
```

```python
ibm_attrition.info()
```

```python
#Checking for missing values
ibm_attrition.isnull().any()
```

```python
#checking the Attrition level
ibm_attrition.Attrition.value_counts()
```

```python
ibm_attrition.DistanceFromHome.value_counts()
```

```python
ibm_attrition.JobRole.value_counts()
```

```python
ibm_attrition.describe()
```

```python
# Breakdown of distance from home by job role and attrition.

dcov_Home=ibm_attrition.groupby(['Attrition', 'JobRole'])['DistanceFromHome'].count()
dcov_Home
```

```python
#Average monthly income by education and attrition.

averageIncome=ibm_attrition.groupby(['Attrition', 'Education'])['MonthlyIncome'].mean()
averageIncome
```

```python
pd.crosstab(ibm_attrition.Attrition, ibm_attrition.JobRole, values=ibm_attrition.DistanceFromHome, aggfunc='mean')
```

```python
pd.crosstab(ibm_attrition.Attrition, ibm_attrition.Education, values=ibm_attrition.MonthlyIncome, aggfunc='mean')
```

```python
#Visualizing the Results
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# Breakdown of Distance from home by Employee Attrition and Job roles
fig=ibm_attrition.groupby(['Attrition', 'JobRole'])['DistanceFromHome'].count().unstack().plot(kind='bar', xlabel='Attrition & JobRole', ylabel="Distance from Home", legend=None, grid=True)

```

```python
#Visualization of Average Monthly Income by Employee attribution and Education. 

fig=ibm_attrition.groupby(['Attrition', 'Education'])['MonthlyIncome'].mean().unstack().plot(kind='barh', ylabel='Attrition and Education', xlabel="Average Monthly Income", grid=True)
```
