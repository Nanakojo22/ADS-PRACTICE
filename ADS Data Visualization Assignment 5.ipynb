#Age distribution

```python
fig = px.histogram(data, x=data['Age'],histnorm='percent', title="Age Distribution")
fig
```

Age Distribution Based on Gender

```python
fig2= px.histogram(data, x ='Age', color='Sex', histnorm='percent', nbins=20, title="Age and Gender Distribution")
fig2
```

#The correlation within the data



```python
data.corr(method ='pearson')
px.imshow(data.corr(method ='pearson'))
```

#Pivot Table 

```python
pivot=pd.pivot_table(data, values='Fare', columns='Pclass')
pivot
```

```python
px.box(data, y='Fare', x='Pclass', title='Distribution of Pclass')
```

```python
Value_count=data.groupby('Survived')['Sex'].value_counts()
Value_count
```

```python
fig = px.histogram(data, x='Survived', color='Sex', barmode="group")
fig
```

```python
#scatter plot of ages to the fare paid by the each passenger based on their gender.

fig = px.scatter(data, x="Age", y="Fare", color="Sex",
                 hover_name="Sex", 
                 log_x=True, size_max=60, title='Age vs Fare Paid By Gender Distribution')
fig.show()
```

#Bubble plot of the ages to the fare paid by each passenger categorizing whether they survived or not.

```python
fig = px.scatter(data, x="Age", y="Fare",
                        size="Pclass", color="Survived",
                 hover_name="Name", log_x=True, size_max=10, title=" Age vs. Fare by Passenger category, Name & Class")
fig.show()
```
